```python
import csv
import mysql.connector

# Database connection details
DB_CONFIG = {
    'host': 'localhost',
    'user': 'root',
    'password': '*********',
    'database': 'phonepe_1'
}

CSV_FILE = r"E:\project\project\csv_file\aggregated_transaction.csv"
TABLE_NAME = 'agg_tran'

def create_table_from_csv_header(cursor, csv_file, table_name):
    with open(csv_file, 'r') as f:
        reader = csv.reader(f)
        header = next(reader) # Get the header row

    # Basic type inference (you might need more sophisticated logic)
    column_definitions = []
    for col_name in header:
        # Simple guess: default to VARCHAR(255)
        # For more robust solutions, read a sample of rows to infer types
        column_definitions.append(f"{col_name} VARCHAR(255)")

    create_table_sql = f"CREATE TABLE IF NOT EXISTS {table_name} ({', '.join(column_definitions)})"
    cursor.execute(create_table_sql)
    print(f"Table '{table_name}' created or already exists.")

def import_csv_to_mysql(csv_file, table_name):
    try:
        conn = mysql.connector.connect(**DB_CONFIG)
        cursor = conn.cursor()

        # Create table based on CSV header (if it doesn't exist)
        create_table_from_csv_header(cursor, csv_file, table_name)

        with open(csv_file, 'r') as f:
            reader = csv.reader(f)
            header = next(reader) # Skip header row as table is already created based on it

            # Prepare the INSERT statement
            placeholders = ', '.join(['%s'] * len(header))
            insert_sql = f"INSERT INTO {table_name} VALUES ({placeholders})"

            for row in reader:
                try:
                    cursor.execute(insert_sql, row)
                except mysql.connector.Error as err:
                    print(f"Error inserting row: {row}. Error: {err}")
                    # Optionally, log the error and continue, or break
                    continue

        conn.commit()
        print("Data imported successfully!")

    except mysql.connector.Error as err:
        print(f"Database error: {err}")
    except FileNotFoundError:
        print(f"Error: CSV file not found at {csv_file}")
    finally:
        if 'conn' in locals() and conn.is_connected():
            cursor.close()
            conn.close()
            print("MySQL connection closed.")

if __name__ == "__main__":
    import_csv_to_mysql(CSV_FILE, TABLE_NAME)
```
